```python
import pandas as pd
import matplotlib.pyplot as plt
illinois =pd.read_csv('/Users/grantflorence/Desktop/Illinois-Data.csv')
```

```python
illinois.head()
```

```python
illinois.columns.tolist()
```

```python
position_pay = illinois.groupby('Position Title')['YTD Gross'].mean()
top_positions = position_pay.sort_values(ascending=False).head(10)
fig, ax = plt.subplots(figsize=(8,6))
ax.barh(top_positions.index, top_positions.values, color='blue')
ax.set_xlabel('Average YTD Gross Pay')
ax.set_title('Top Paying Positions in the State of Illinois based on YTD Gross Pay')
ax.invert_yaxis()
plt.show()
```
